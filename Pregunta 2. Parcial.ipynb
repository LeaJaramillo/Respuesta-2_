**2. Ejercicio:**

Una cadena de televisión ha presentado un nuevo programa para la franja de las 11 a las 15 horas. El share o porcentaje de audiencia de la primera emisión vino dado por la siguiente función, donde S(t) representa el share en el tiempo t, en horas. Para que el programa siga emitiéndose el share ha tenido que alcanzar, en algún momento el 30%

S(t)=−t3+36t2−420t+1596;11≤t≤15


Indica cuándo creció el share y cuándo decreció ¿El programa seguirá emitiéndose?
Dibuja la gráfica del share.

```python
# Importar las herramientas necesarias
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
from sympy.abc import x

init_printing()

# Definir la función con la cual operaremos
# Esta parte se debe ajustar de acuerdo a la función
f = (1/6) * x**3 - (1/2) * x**2 - (3/2) * x + (1/6)

def maxminf(f):
    """Calcula los máximos y mínimos de una función f(x) y muestra el gráfico."""
    df = diff(f, x)  # 1era derivada
    d2f = diff(f, x, 2)  # 2da derivada

    # Encontrar los puntos críticos
    pcs = solve(Eq(df, 0))
    
    if not pcs:
        print("No se encontraron puntos críticos.")
        return

    # Evaluar cada punto crítico y determinar si es Máx o Mín
    maximos = []
    minimos = []

    for p in pcs:
        f_p = f.subs(x, p)  # Valor de la función en el punto crítico
        segunda_eval = d2f.subs(x, p)

        if segunda_eval > 0:
            tipo = "Mínimo"
            minimos.append((p, f_p))
        elif segunda_eval < 0:
            tipo = "Máximo"
            maximos.append((p, f_p))
        else:
            tipo = "Indefinido"

        print(f"x = {p:.2f}, f(x) = {f_p:.2f} ({tipo})")

    # Graficar la función
    x_vals = np.linspace(-10, 10, 400)
    f_lambdified = lambdify(x, f, modules=['numpy'])
    y_vals = f_lambdified(x_vals)

    plt.figure(figsize=(6, 4))
    plt.plot(x_vals, y_vals, label='f(x)', color='green')

    # Resaltar los puntos críticos
    for p, f_p in maximos:
        plt.plot(float(p), float(f_p), 'bo')  # Máximos en azul
        plt.text(float(p), float(f_p), f"({p:.2f}, {f_p:.2f})", fontsize=10, color='blue')

    for p, f_p in minimos:
        plt.plot(float(p), float(f_p), 'ro')  # Mínimos en rojo
        plt.text(float(p), float(f_p), f"({p:.2f}, {f_p:.2f})", fontsize=10, color='red')

    # Configurar el gráfico
    plt.axhline(0, color='black', linewidth=0.8)
    plt.axvline(0, color='black', linewidth=0.8)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Gráfico de la función y sus extremos')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(loc='best')
    plt.show()

# Llamar a la función para que el usuario interactúe
maxminf(f)
```
